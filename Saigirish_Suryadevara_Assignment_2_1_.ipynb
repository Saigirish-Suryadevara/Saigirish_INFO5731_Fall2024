Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

```python
# Write code for each of the sub parts with proper comments.
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# downloading required nltk data
nltk.download('wordnet')
nltk.download('stopwords')

csv_df = pd.read_csv('Movie_Review.csv')

#1. Remove noise, special characters & punctuations

def clean_noise(text):
  return re.sub(r'[^A-Za-z\s]', '', text)

df['cleaned_Data_noise'] = df['Review'].apply(clean_noise)
print("After removing special characters and punctuation:\n")
print(df[['Review','cleaned_Data_noise']].head(),"\n")


```

```python
#2 Remove Numbers
def clean_numbers(text):
  return re.sub(r'\d+', '', text)

df['cleaned_Data_numbers'] = df['cleaned_Data_noise'].apply(clean_numbers)
print("Data after removing numbers:\n")
print(df[['cleaned_Data_noise','cleaned_Data_numbers']].head(),"\n")
```

```python
#3 Remove Stopwords
StopWords = set(stopwords.words('english'))
def clean_stopwords(text):
  word = text.split()
  cleansed_word = [w for w in word if w.lower() not in StopWords]
  return ' '.join(cleansed_word)

df['cleaned_Data_stopwords'] = df['cleaned_Data_numbers'].apply(clean_stopwords)
print("Data after removing stop words:\n")
print(df[['cleaned_Data_numbers','cleaned_Data_stopwords']].head(),"\n")

```

```python
#4 Changing all letters to lower case
def clean_lowercase(text):
  return text.lower()

df['cleaned_Data_lowercase'] = df['cleaned_Data_stopwords'].apply(clean_lowercase)
print("Data after changing all letters to lower case:\n")
print(df[['cleaned_Data_stopwords','cleaned_Data_lowercase']].head(),"\n")
```

```python
#5 Stemming
stemming = PorterStemmer()
def clean_stemming(text):
  word = text.split()
  word_stemmed = [stemming.stem(w) for w in word]
  return ' '.join(word_stemmed)

df['cleaned_Data_stemming'] = df['cleaned_Data_lowercase'].apply(clean_stemming)
print("Data after stemming:\n")
print(df[['cleaned_Data_lowercase','cleaned_Data_stemming']].head(),"\n")
```

```python
#6 Lemmatization
lemmatizing = WordNetLemmatizer()
def clean_lemmatization(text):
  word = text.split()
  word_lemmatized = [lemmatizing.lemmatize(w) for w in word]
  return ' '.join(word_lemmatized)

df['cleaned_Data_lemmatizing'] = df['cleaned_Data_stemming'].apply(clean_lemmatization)
print("Data after lemmatization:\n")
print(df[['cleaned_Data_stemming','cleaned_Data_lemmatizing']].head(),"\n")


# Saving cleaned data after all above steps to a new csv file

df.to_csv('Complete_Cleaned_Movie_Review.csv', index= False)
print("\n Data Saved to Complete_Cleaned_Movie_Review.csv")


```

# Question 3 (30 points)

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(1) **Parts of Speech (POS) Tagging:** Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
# Your code here

import pandas as pd
import spacy
from collections import Counter
import nltk
from nltk.tree import Tree

# Downloading required nltk resources
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('words')
nltk.download('maxent_ne_chunker')

#loading spacy model for eng
nlp = spacy.load('en_core_web_sm')

df = pd.read_csv('Complete_Cleaned_Movie_Review.csv')

# POS tagging

def pos_tag(text):
  docmnt = nlp(text)
  count_pos = Counter([token.pos_ for token in docmnt])
  return count_pos

all_noun = all_verb = all_adverb = all_adjective = 0


# taking lemmatised and clean data from question 2nd from here when ever required
for i,review in enumerate(df['cleaned_Data_lemmatizing']):
  count_pos = pos_tag(review)
  all_noun += count_pos.get('NOUN', 0)
  all_verb += count_pos.get('VERB', 0)
  all_adverb += count_pos.get('ADV', 0)
  all_adjective += count_pos.get('ADJ', 0)


# print all count of each pos category
print("Count of all POS tag Categories:\n")
print(f"\nCount of Nouns: {all_noun}")
print(f"\nCount of Verbs: {all_verb}")
print(f"\nCount of Adverbs: {all_adverb}")
print(f"\nCount of Adjectives: {all_adjective}")


#2 Constituency parsing & Dependecny parsing
print("\n\nConstituency parsing for first review:\n")
example_review = df['cleaned_Data_lemmatizing'].iloc[0]

# performing Constituency parsing
token = nltk.word_tokenize(example_review)
tag = nltk.pos_tag(token)
ne_chunk = nltk.ne_chunk(tag)

# printing the constituency tree
print("\n Constituency tree:")
print(Tree.fromstring(str(ne_chunk)).pretty_print())

"""Constituency parsing splits down sentence to it's grammatic sense and depicts how words group up to constitute phrase.
   This explains the heirarchial relationship among the phrases, showing overall structure.
   """

# Dependency Parsing
print("\nDependency parsing for first review")
docmnt = nlp(example_review)
print(f"Review: '{example_review}'\n")

#print dependency tree
for token in docmnt:
  print(f"{token.text}: {token.dep_} <- {token.head.text}")

""" Dependency Parsing shows grammatic relation among each individual words, where each word points towards it head word.
    It shows how each word is dependant on other and explaine their relation and roles in the structure."""

#3. Named entitiy recognition
def get_entity(text):
  docmnt = nlp(text)
  entity = [(e.text, e.label_) for e in docmnt.ents]
  return entity

count_entity = Counter()
# Applying NER for each review
for i,review in enumerate(df['cleaned_Data_lemmatizing']):
  entity = get_entity(review)
  count_entity.update([e[1] for e in entity])

#printing entities count
print("\n Count of Entities:")
for type_entity,c in count_entity.items():
  print(f"{type_entity}: {c}")





```
