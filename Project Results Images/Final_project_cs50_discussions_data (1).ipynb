```python
import praw
import pandas as pd

reddit = praw.Reddit(user_agent=True,
                     client_id='******',# use your client id
                     client_secret='*********', # use your client secret
                     username='******', # use your client username
                     password='*******') # use your client password


keywords = ['data science', 'python','machine learning','c++', 'java']
data = []
target_comments = 6000

for k in keywords:
  posts = reddit.subreddit('cs50').search(k, limit = None)
  for p in posts:
    post_id = p.id
    post_title = p.title
    post_score = p.score
    post_url = p.url
    Keyword_search = k

    p.comments.replace_more(limit=0)
    for c in p.comments.list():
      if len(data)>=target_comments:
        break

      comment_id = c.id
      parent_id = c.parent_id
      comment_depth = c.depth
      comment_body = c.body
      comment_likes = c.score

      start_discussion = 1 if comment_depth == 0 else 0

      data.append([post_id,post_title,post_score,post_url,Keyword_search,comment_id,parent_id,
                   comment_depth,comment_body,comment_likes,start_discussion])
    if len(data)>=target_comments:
      break
  if len(data)>=target_comments:
    break

df = pd.DataFrame(data, columns=['post_id','post_title','post_score','post_url','Keyword_search','comment_id','parent_id',
                   'comment_depth','comment_body','comment_likes','start_discussion'])

df.to_csv('cs50 Discussions.csv', index=False, encoding='utf-8')
print(f"{len(data)} comments have been retrieved ans saved to csv")

```

```python
import pandas as pd
import re


df = pd.read_csv('/content/cs50 Discussions (1).csv')


def clean_text(text):

    text = text.lower()


    text = re.sub(r'http\S+|www.\S+', '', text)


    text = re.sub(r'[^a-z\s]', '', text)


    text = re.sub(r'\s+', ' ', text).strip()

    return text


df['comment_body_cleaned'] = df['comment_body'].astype(str).apply(clean_text)
df['post_title_cleaned'] = df['post_title'].astype(str).apply(clean_text)


df.drop_duplicates(subset=['comment_body_cleaned'], inplace=True)
df.reset_index(drop=True, inplace=True)


df.to_csv('cleaned_reddit_comments.csv', index=False, encoding='utf-8')

print("Cleaning completed. The cleaned data has been saved to 'cleaned_reddit_comments.csv'.")
```

```python
import pandas as pd
from statsmodels.stats.inter_rater import fleiss_kappa

# Load the data from the Excel file
file_path = '/content/1500 Labeled Posts.csv'
data = pd.read_csv(file_path, usecols=['Girish', 'Sreyesh', 'Ganesh','Divya','Rithvik','Krishna'])

# Create a new DataFrame to store the counts of each rating per subject
ratings_counts = pd.DataFrame(0, index=data.index, columns=[0,1, 2, 3, 4, 5])

# Count the ratings for each subject
for col in data.columns:
    counts = pd.get_dummies(data[col])
    ratings_counts = ratings_counts.add(counts, fill_value=0)

# Ensure the DataFrame has the correct dtype for Fleiss' kappa computation
ratings_counts = ratings_counts.astype(int)

# Compute Fleiss' kappa
kappa = fleiss_kappa(ratings_counts.values, method='fleiss')

print("Fleiss' kappa:", kappa)
```
